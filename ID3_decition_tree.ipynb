```python
import pandas as pd
import math
df = pd.read_csv('drive/MyDrive/golf-dataset.csv')
print(df) 
```

```python
#taking values to list
playgolf = df['Play Golf'].tolist()  #class atrribute
outlook =  df['Outlook'].tolist()
humidity =  df['Humidity'].tolist()
windy =  df['Windy'].tolist()
temp =  df['Temp'].tolist()
```

```python
def find_entropy(atr_yes,atr_no,atr_len):
  entropy = 0
  if atr_yes>=1 and atr_no >=1:
    entropy = -(atr_yes/atr_len)*math.log2(atr_yes/atr_len)-(atr_no/atr_len)*math.log2(atr_no/atr_len)
  elif atr_yes>=1:
    entropy = -(atr_yes/atr_len)*math.log2(atr_yes/atr_len)
  elif atr_no>=1:
    entropy = -(atr_no/atr_len)*math.log2(atr_no/atr_len)
  else:
    entropy = 0
  return entropy

#calculating entropy of playgolfs
p_yes= playgolf.count("Yes")
p_no= playgolf.count("No")
p_len =len(playgolf)
p_entropy = find_entropy(p_yes,p_no,p_len)
print("playgolf entropy",p_entropy)
```

```python
#calculating gain of outlook
#startimg with calculating entropy of sunny
total_sunny = outlook.count("Sunny")
s_yes=0
s_no =0
for i in range(len(outlook)):
  if outlook[i]=="Sunny":
    if playgolf[i]=="Yes":
      s_yes+=1
    else:
      s_no+=1
s_entropy = find_entropy(s_yes,s_no,total_sunny)
print("sunny entropy",s_entropy)

#entropy of overcast
total_overcast = outlook.count("Overcast")
o_yes=0
o_no =0
for i in range(len(outlook)):
  if outlook[i]=="Overcast":
    if playgolf[i]=="Yes":
      o_yes+=1
    else:
      o_no+=1
o_entropy = find_entropy(o_yes,o_no,total_overcast)
print("Overcast entropy",o_entropy)

#entropy of Rainy
total_rain = outlook.count("Rainy")
r_yes=0
r_no =0
for i in range(len(outlook)):
  if outlook[i]=="Rainy":
    if playgolf[i]=="Yes":
      r_yes+=1
    else:
      r_no+=1
r_entropy = find_entropy(r_yes,r_no,total_rain)
print("rain entropy",r_entropy)


o_gain = p_entropy - (total_sunny/p_len)*(s_entropy) - (total_overcast/p_len)*(o_entropy) -(total_rain/p_len)*(r_entropy)
print("gain of outlook",o_gain)
```

```python
#temp column
#calculating gain of temp
#startimg with calculating entropy of cool
total_cool = temp.count("Cool")
c_yes=0
c_no =0
for i in range(len(temp)):
  if temp[i]=="Cool":
    if playgolf[i]=="Yes":
      c_yes+=1
    else:
      c_no+=1
c_entropy = find_entropy(c_yes,c_no,total_cool)
print("Cool entropy",c_entropy)

#startimg with calculating entropy of Mild
total_mild = temp.count("Mild")
m_yes=0
m_no =0
for i in range(len(temp)):
  if temp[i]=="Mild":
    if playgolf[i]=="Yes":
      m_yes+=1
    else:
      m_no+=1
m_entropy = find_entropy(m_yes,m_no,total_mild)
print("Mild entropy",m_entropy)

#startimg with calculating entropy of hot
total_hot = temp.count("Hot")
h_yes=0
h_no =0
for i in range(len(temp)):
  if temp[i]=="Hot":
    if playgolf[i]=="Yes":
      h_yes+=1
    else:
      h_no+=1
h_entropy = find_entropy(h_yes,h_no,total_hot)
print("Hot entropy",h_entropy)

t_gain = p_entropy - (total_cool/p_len)*(c_entropy) - (total_mild/p_len)*(m_entropy) -(total_hot/p_len)*(h_entropy)
print("gain of temp",t_gain)
```

```python
#humidity column
#calculating gain of humidity
#startimg with calculating entropy of high
total_high = humidity.count("High")
h_yes=0
h_no =0
for i in range(len(humidity)):
  if humidity[i]=="High":
    if playgolf[i]=="Yes":
      h_yes+=1
    else:
      h_no+=1
h_entropy = find_entropy(h_yes,h_no,total_high)
print("High entropy",h_entropy)

#startimg with calculating entropy of normal
total_normal = humidity.count("Normal")
n_yes=0
n_no =0
for i in range(len(humidity)):
  if humidity[i]=="Normal":
    if playgolf[i]=="Yes":
      n_yes+=1
    else:
      n_no+=1
n_entropy = find_entropy(n_yes,n_no,total_normal)
print("Normal entropy",n_entropy)


h_gain = p_entropy - (total_high/p_len)*(h_entropy) - (total_normal/p_len)*(n_entropy)
print("gain of humidity",h_gain)
```

```python
#windy column
#calculating gain of windy
#startimg with calculating entropy of high
total_true = windy.count(True)
t_yes=0
t_no =0
for i in range(len(windy)):
  if windy[i]==True:
    if playgolf[i]=="Yes":
      t_yes+=1
    else:
      t_no+=1

t_entropy = find_entropy(t_yes,t_no,total_true)
print("True entropy",t_entropy)

#startimg with calculating entropy of normal
total_false = windy.count(False)
f_yes=0
f_no =0
for i in range(len(windy)):
  if windy[i]==False:
    if playgolf[i]=="Yes":
      f_yes+=1
    else:
      f_no+=1
f_entropy = find_entropy(f_yes,f_no,total_false)
print("False entropy",f_entropy)


w_gain = p_entropy - (total_true/p_len)*(t_entropy) - (total_false/p_len)*(f_entropy)
print("gain of windy",w_gain)
```

```python
print("gain of outlook",o_gain)
print("gain of temp",t_gain)
print("gain of humidity",h_gain)
print("gain of wind",w_gain)
root1 = 0
root = ""
if o_gain > max(t_gain, h_gain,w_gain):
  root1 =o_gain
  root = "Outlook"
  print(" Root = Outlook")
elif t_gain > max(o_gain, h_gain,w_gain):
  root1 =t_gain
  root = "Temp"
  print(" Root = Temp")
elif h_gain > max(o_gain, t_gain,w_gain):
  root1 =h_gain
  root = "Humidity"
  print(" Root = Humidity")
elif w_gain > max(o_gain, h_gain,t_gain):
  root1 =w_gain
  root = "Windy"
  print(" Root = Windy")
```
